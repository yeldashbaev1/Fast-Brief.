```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading images
img1 = cv2.imread("face.jpeg")
img2 = cv2.imread("faceQS.png")

# Converting BGR to RGB
img11 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

# Converting RGB to GRAY
img1 = cv2.cvtColor(img1,cv2.COLOR_RGB2GRAY)
img2 = cv2.cvtColor(img2,cv2.COLOR_RGB2GRAY)

# Taking object from ORB
orb = cv2.ORB_create(500,2.0)

# Detect and Compute
key1,des1 = orb.detectAndCompute(img1,None)
key2,des2 = orb.detectAndCompute(img2,None)

# Resizing if necessary
plt.figure(figsize=(10,10))

# Taking object from Brief matcher function
bf = cv2.BFMatcher(crossCheck=True)

# Finding similarities by distance
similar = bf.match(des1,des2)

# Sorting by distance
similar = sorted(similar,key=lambda x: x.distance)

# Taking result and drawing matches
result = cv2.drawMatches(img1,key1,img2,key2,similar[:30],img2,flags=2)

# Showing result
plt.imshow(result)
```
